```python
import pandas as pd
```

```python
df=pd.read_csv("enriched_diet_dataset.csv")
```

```python
df
```

```python
# One-hot encode categorical variables
df_encoded = pd.get_dummies(df, columns=['Gender', 'Activity Level', 'Goal'], drop_first=True)
```

```python
df[['Gender', 'Activity Level', 'Goal']]
```

```python
# Input features (X)
X = df_encoded[['Age', 'Height', 'Weight', 'Days'] + 
               [col for col in df_encoded.columns if col.startswith('Gender_') or 
                                                    col.startswith('Activity Level_') or 
                                                    col.startswith('Goal_')]]

# Target variables (Y)
Y = df_encoded[['Target Calories', 'Protein (g)', 'Carbs (g)', 'Fats (g)']]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```
