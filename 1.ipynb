```python
import pandas as pd
import numpy as np

df = pd.read_excel("Min_macros_for_height_and_weight.xlsx")
```

```python
df['Food Energy (Calories/day)'] = df['Food Energy (Calories/day)'].str.replace(',', '').astype(float)

np.random.seed(42)  

df['Age'] = np.random.randint(18, 61, size=len(df))
df['Gender'] = np.random.choice(['Male', 'Female'], size=len(df))
df['Activity Level'] = np.random.choice(['Sedentary', 'Lightly active', 'Moderately active', 'Very active'], size=len(df))
df['Goal'] = np.random.choice(['Lose', 'Gain', 'Maintain'], size=len(df))
df['Days'] = np.random.randint(30, 121, size=len(df))
```

```python
df = df.drop(['Unnamed: 0'], axis=1)
```

```python
activity_multipliers = {
    'Sedentary': 1.2,
    'Lightly active': 1.375,
    'Moderately active': 1.55,
    'Very active': 1.725
}

def calculate_bmr(row):
    weight = row['Weight']
    height = row['Height']
    age = row['Age']
    gender = row['Gender']
    
    if gender == 'Male':
        return 10 * weight + 6.25 * height - 5 * age + 5
    else:
        return 10 * weight + 6.25 * height - 5 * age - 161
```

```python
df['BMR'] = df.apply(calculate_bmr, axis=1)
df['TDEE'] = df['BMR'] * df['Activity Level'].map(activity_multipliers)
```

```python
def calculate_target_calories(row):
    if row['Goal'] == 'Lose':
        return row['TDEE'] - 500
    elif row['Goal'] == 'Gain':
        return row['TDEE'] + 500
    else:
        return row['TDEE']

df['Target Calories'] = df.apply(calculate_target_calories, axis=1)
```

```python
df[['Age', 'Gender', 'Activity Level', 'Goal', 'BMR', 'TDEE', 'Target Calories']].head()
```

```python
# Calculate macro calories
df['Protein Calories'] = df['Target Calories'] * 0.25
df['Carb Calories']    = df['Target Calories'] * 0.45
df['Fat Calories']     = df['Target Calories'] * 0.30

# Convert calories to grams
df['Protein (g)'] = df['Protein Calories'] / 4
df['Carbs (g)']   = df['Carb Calories'] / 4
df['Fats (g)']    = df['Fat Calories'] / 9

# Round for cleaner display
df[['Protein (g)', 'Carbs (g)', 'Fats (g)']] = df[['Protein (g)', 'Carbs (g)', 'Fats (g)']].round(1)
```
